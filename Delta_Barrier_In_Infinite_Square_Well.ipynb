```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as sc
import math
from scipy.optimize import fsolve

a=1;n=3
total=100
x=np.linspace(-a,a,total)
alpha=(((sc.hbar)**2))/(sc.m_e*a)
eps=(((sc.hbar)**2))/(sc.m_e*alpha*a)
guess=(n*math.pi)/2
def trans_eqn(z):
  return np.tan(z)+eps*z
z=fsolve(trans_eqn,guess)
coeff=(a)*((1-(np.sin(2*z)/(2*z)))+eps*(1-np.cos(2*z))+eps*eps*z*z*(1+(np.sin(2*z)/(2*z))))
A=1/math.sqrt(coeff)
B=eps*z*A
C=-A*np.tan(z)
def even_psi_2(x):
  return A*np.sin((z*x)/a)+B*np.cos((z*x)/a)
def even_psi_1(x):
  return -A*np.sin((z*x)/a)+B*np.cos((z*x)/a)

y_even=[]
for i in range(0,int(total/2)):
  y_even.append(even_psi_1(x[i]))
for i in range(int(total/2),total):
  y_even.append(even_psi_2(x[i]))
Z=np.linspace(-10,10,100)
plt.plot(x,y_even)
```
